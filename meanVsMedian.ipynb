This code was used to create the charts and images for a linkedin post comparing the mean and median as measures of central tendency. 



```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import lognorm
```

```python
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")
```

```python
# Parameters for the normal distribution
np.random.seed(99)
mean = 50 
std_dev = 5 
num_samples = 50


normData = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
normX = np.full(num_samples, 0.15)
normMean = np.mean(normData)
normMedian = np.median(normData)

print(normMean)
print(normMedian)

fig, ax = plt.subplots(
    figsize=(10,6.8)
)

ax.set_xlim(0, 0.178)

ax.scatter(normX, normData, alpha=0.25)
ax.hist(normData, bins=30, density=True, alpha=0.5, color='steelblue', orientation='horizontal')
sns.kdeplot(y=normData, ax=ax, color='#232323', alpha=0.5, linewidth=2)

ax.plot([0.13,.145], [normMean, normMean], color='#232323', linestyle='-')
ax.annotate('Mean', xy=(0.1, normMean), xytext=(0.145, normMean + 0.75),
             fontsize=20, color='#232323', ha='right')
ax.annotate(round(normMean,1), xy=(0.13, normMean), xytext=(0.145, normMean - 2),
             fontsize=20, color='#232323', ha='right')

ax.plot([0.155,0.17], [normMedian, normMedian], color='#232323', linestyle='-')
ax.annotate('Median', xy=(0.155, normMedian), xytext=(0.155, normMedian + 0.75),
             fontsize=20, color='#232323', ha='left')
ax.annotate(round(normMedian,1), xy=(0.155, normMedian), xytext=(0.155, normMedian - 2),
             fontsize=20, color='#232323', ha='left')

ax.set_title('Normally distributed data', fontsize=22)

ax.tick_params(axis='y', labelsize=22)

ax.xaxis.set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()

plt.savefig('normDist.png', dpi=192)

plt.show()

```

```python
# Parameters for the skewed distribution
np.random.seed(99)

shape = 0.75 
loc = 0  
scale = np.exp(4)  # scale parameter (exp(mu))

# Generate the data
num_samples = 50
skewData = lognorm.rvs(s=shape, loc=loc, scale=scale, size=num_samples)
skewX = np.full(num_samples, .0175)

skewMean = np.mean(skewData)
skewMedian = np.median(skewData)
# Plot the data to visualize the distribution
print(skewMean)
print(skewMedian)

# sns.set_style("white")
sns.despine(right = True, top = True, bottom = True)

fig, ax = plt.subplots(
    figsize=(10,6.8)
)

ax.set_xlim(0, 0.02)

ax.scatter(skewX, skewData, alpha=0.25, c='#8fbc8f')
ax.hist(skewData, bins=30, density=True, alpha=0.5, color='#8fbc8f', orientation='horizontal')
sns.kdeplot(y=skewData, ax=ax, color='#232323', alpha=0.5, linewidth=2)

ax.plot([0.0155,.017], [skewMean, skewMean], color='#232323', linestyle='-')
ax.annotate('Mean', xy=(0.0155, skewMean), xytext=(0.017, skewMean + 5),
  fontsize=20, color='#232323', ha='right')
ax.annotate(round(skewMean,1), xy=(0.0155, skewMean), xytext=(0.017, skewMean - 20),
  fontsize=20, color='#a22222', ha='right')

ax.plot([0.018,0.0195], [skewMedian, skewMedian], color='#232323', linestyle='-')
ax.annotate('Median', xy=(0.018, skewMedian), xytext=(0.018, skewMedian + 5),
  fontsize=20, color='#232323', ha='left')
ax.annotate(round(skewMedian,1), xy=(0.018, skewMedian), xytext=(0.018, skewMedian - 20),
  fontsize=20, color='#232323', ha='left')

ax.set_title('positively skewed data', fontsize=22)

ax.tick_params(axis='y', labelsize=22)

ax.xaxis.set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()

plt.savefig('skewDist.png', dpi=192)

plt.show()
```

```python
# Parameters for the normal distribution with an outliner
np.random.seed(99)
mean = 50 
std_dev = 5 
num_samples = 49


outlData = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
outlX = np.full(num_samples + 1, 0.015)

outlierVal = 1000

outlData = np.append(outlData,  outlierVal)


outlMean = np.mean(outlData)
outlMedian = np.median(outlData)


print(outlMean)
print(outlMedian)

fig, ax = plt.subplots(
    figsize=(10,6.8)
)

ax.set_xlim(0, 0.02)

ax.scatter(outlX, outlData, alpha=0.25, color='#ff8c00')
ax.hist(outlData, bins=30, density=True, alpha=0.5, color='#ff8c00', orientation='horizontal')
sns.kdeplot(y=outlData, ax=ax, color='#232323', alpha=0.5, linewidth=2)

ax.plot([0.013,.0145], [outlMean, outlMean], color='#232323', linestyle='-')
ax.annotate('Mean', xy=(0.013, outlMean), xytext=(0.0145, outlMean + 20),
  fontsize=20, color='#232323', ha='right')
ax.annotate(round(outlMean,1), xy=(0.013, outlMean), xytext=(0.0145, outlMean - 85),
fontsize=20, color='#a22222', ha='right')

ax.plot([0.016, 0.0175], [outlMedian, outlMedian], color='#232323', linestyle='-')
ax.annotate('Median', xy=(0.016, outlMedian), xytext=(0.016, outlMedian + 20),
  fontsize=20, color='#232323', ha='left')
ax.annotate(round(outlMedian,1), xy=(0.016, outlMedian), xytext=(0.016, outlMedian - 85),
  fontsize=20, color='#232323', ha='left')

ax.set_title('normally distributed data with an outlier', fontsize=22)

ax.tick_params(axis='y', labelsize=22)

ax.xaxis.set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()

plt.savefig('oultDist.png', dpi=192)

plt.show()


```
